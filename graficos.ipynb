```python
import pandas as pd
import matplotlib.pyplot as plt

# Defina o nome do seu arquivo de dados
caminho_arquivo = 'results.csv'

# Tenta carregar os dados do arquivo CSV
try:
    # CORRE√á√ÉO: Especificamos a codifica√ß√£o 'utf-16' usada pelo PowerShell
    df = pd.read_csv(caminho_arquivo, encoding='utf-16')
except FileNotFoundError:
    print(f"Erro: O arquivo '{caminho_arquivo}' n√£o foi encontrado.")
    print("Por favor, certifique-se de que este script Python est√° na MESMA PASTA que o seu arquivo 'results.csv'.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
    exit()

# Cria a figura e os eixos para o gr√°fico
plt.figure(figsize=(12, 8))

# Plota os dados de ambos os m√©todos
plt.plot(df['Tamanho_N'], df['Tempo_Linhas_s'], marker='o', linestyle='-', color='blue', label='Acesso por Linhas (Cache-Friendly)')
plt.plot(df['Tamanho_N'], df['Tempo_Colunas_s'], marker='x', linestyle='--', color='red', label='Acesso por Colunas (Cache-Unfriendly)')

# Configura as escalas dos eixos para melhor visualiza√ß√£o
plt.xscale('log', base=2)
plt.yscale('log')

# Adiciona t√≠tulos e r√≥tulos
plt.title('Impacto do Padr√£o de Acesso √† Mem√≥ria na Performance (Escala Log-Log)', fontsize=16)
plt.xlabel('Tamanho da Matriz (N)', fontsize=12)
plt.ylabel('Tempo M√©dio de Execu√ß√£o (segundos)', fontsize=12)

# Adiciona uma legenda
plt.legend(fontsize=11)

# Adiciona marcadores no eixo X
plt.xticks(df['Tamanho_N'], df['Tamanho_N'], rotation=45)

# Adiciona uma grade
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Ajusta o layout
plt.tight_layout()

# Salva o gr√°fico em um arquivo de imagem
nome_grafico = 'grafico_performance_cache.png'
plt.savefig(nome_grafico)

print(f"Gr√°fico '{nome_grafico}' gerado com sucesso!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Defina o nome do seu arquivo de dados
caminho_arquivo = 'results.csv'

# Tenta carregar os dados do arquivo CSV
try:
    # Usando a codifica√ß√£o 'utf-16' que o PowerShell gera
    df = pd.read_csv(caminho_arquivo, encoding='utf-16')
except FileNotFoundError:
    print(f"Erro: O arquivo '{caminho_arquivo}' n√£o foi encontrado.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
    exit()

# Cria a figura e os eixos para o gr√°fico
plt.figure(figsize=(10, 6))

# Plota o Fator de Lentid√£o em fun√ß√£o do Tamanho da Matriz
plt.plot(df['Tamanho_N'], df['Fator_Lentidao'], marker='o', linestyle='-', color='green', label='Fator de Lentid√£o (Colunas / Linhas)')

# Configura a escala do eixo X para log, para ver melhor a progress√£o
plt.xscale('log', base=2)
# A escala Y ser√° linear para vermos o crescimento do fator diretamente

# Adiciona t√≠tulos e r√≥tulos
plt.title('Crescimento do Fator de Lentid√£o vs. Tamanho da Matriz', fontsize=16)
plt.xlabel('Tamanho da Matriz (N)', fontsize=12)
plt.ylabel('Fator de Lentid√£o (x vezes mais lento)', fontsize=12)

# Adiciona uma legenda
plt.legend(fontsize=11)

# Adiciona marcadores no eixo X
plt.xticks(df['Tamanho_N'], df['Tamanho_N'], rotation=45)

# Adiciona uma grade
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Adiciona uma linha horizontal em y=1.0 para refer√™ncia de "sem diferen√ßa"
plt.axhline(y=1.0, color='gray', linestyle=':', linewidth=1.5)

# Ajusta o layout
plt.tight_layout()

# Salva o gr√°fico em um arquivo de imagem
nome_grafico = 'grafico_fator_lentidao.png'
plt.savefig(nome_grafico)

print(f"Gr√°fico '{nome_grafico}' gerado com sucesso!")
```
